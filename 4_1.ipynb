```julia
using Plots
a = 0.000133
b = 0.143

dt = 0.01
day = 50
nIter = Int(day/dt)

S = zeros(nIter)
I = zeros(nIter)
R = zeros(nIter)
S[1] = 14990
I[1] = 10
R[1] = 0

for n in 1:(nIter-1)
    S[n+1] = S[n] + (-a*S[n]*I[n])*dt
    I[n+1] = I[n] + (a*S[n]*I[n] - b*I[n])*dt
    R[n+1] = R[n] + (b*I[n])*dt
end

plot(1:nIter, S)
plot!(1:nIter, I)
plot!(1:nIter, R)
```
